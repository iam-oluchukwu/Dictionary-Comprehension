# Student Final Exam Requirement (Dictionary Comprehension)

## Objective
This program calculates how many points each student needs in the final exam to pass.

## Method
- Calculate average of two scores
- Use dictionary comprehension
- Find required final score to reach 50

```python
students = [
    {'name': 'Ally', 'score': [42, 48]},
    {'name': 'Betty', 'score': [68, 45]},
    {'name': 'Kirby', 'score': [90, 85]},
    {'name': 'Samuel', 'score': [35, 55]}
]

# Calculate required final score
result = {
    s['name']: max(0, 50 * 3 - sum(s['score']))
    for s in students
}

print(result)
```
